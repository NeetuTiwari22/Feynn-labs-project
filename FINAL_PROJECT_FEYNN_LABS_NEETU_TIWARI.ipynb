```python
from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['STATCD']
  ys = series['TOTPOPULAT']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = _df_12.sort_values('STATCD', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('DISTNAME')):
  _plot_series(series, series_name, i)
  fig.legend(title='DISTNAME', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('STATCD')
_ = plt.ylabel('TOTPOPULAT')
```

```python
import pandas as pd
df = pd.read_csv('/content/engineering colleges in India.csv')
df.head()
```

```python
# @title Country vs College Type

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['College Type'].value_counts()
    for x_label, grp in df.groupby('Country')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Country')
_ = plt.ylabel('College Type')
```
