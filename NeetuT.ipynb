```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Define the path to your CSV file using a raw string
csv_file_path ='/content/drive/MyDrive/mcdonalds.csv'

# Load data from the CSV file using pandas
df = pd.read_csv(csv_file_path)

# Assuming the relevant columns are the first 11 columns
MD_x = np.array(df.iloc[:, 0:11] == "Yes", dtype=int)

# Perform PCA
pca = PCA(n_components=11)
pca.fit(MD_x)

# Get the standard deviations of each principal component
std_devs = np.sqrt(pca.explained_variance_)
print("Standard deviations:", np.round(std_devs, 1))

# Get the rotation matrix (components)
rotation = pca.components_.T
print("Rotation matrix (n x k) = (11 x 11):")
rotation_df = pd.DataFrame(rotation, index=df.columns[:11], columns=[f"PC{i+1}" for i in range(11)])
print(rotation_df.round(2))

# Plotting the first two principal components
transformed_data = pca.transform(MD_x)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1], color='grey')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of McDonalds Dataset')
plt.grid(True)
plt.show()

# Optional: Projecting axes
# This step involves more detailed plotting, which can be done using matplotlib if needed.
# Below is a simplified example for axis projection visualization.

fig, ax = plt.subplots()
scatter = ax.scatter(transformed_data[:, 0], transformed_data[:, 1], color='grey')

# Draw the principal component axes
for i in range(rotation.shape[1]):
    ax.arrow(0, 0, rotation[0, i], rotation[1, i], color='r', width=0.01)
    ax.text(rotation[0, i], rotation[1, i], f"PC{i+1}", color='r')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of McDonalds Dataset with Axes')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from scipy.stats import bootstrap
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Define the path to your CSV file using a raw string
csv_file_path ='/content/drive/MyDrive/mcdonalds.csv'

# Load data from the CSV file using pandas
df = pd.read_csv(csv_file_path)

# Assuming the relevant columns are the first 11 columns
MD_x = np.array(df.iloc[:, 0:11] == "Yes", dtype=int)

# Perform PCA
pca = PCA(n_components=11)
MD_pca = pca.fit_transform(MD_x)

# Plotting the PCA projection
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], color='grey')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Projection')
plt.grid(True)
plt.show()

# Perform k-means clustering for k=2 to k=8
clusters = range(2, 9)
inertias = []
best_kmeans_models = {}

for k in clusters:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=1234)
    kmeans.fit(MD_x)
    inertias.append(kmeans.inertia_)
    best_kmeans_models[k] = kmeans

# Plot the inertia to determine the optimal number of clusters
plt.plot(clusters, inertias, marker='o')
plt.xlabel('Number of segments')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()

# Bootstrapping and adjusted Rand index for k=2 to k=8
boot_results = {k: [] for k in clusters}
n_bootstraps = 100

for k in clusters:
    for _ in range(n_bootstraps):
        # Bootstrap sampling with replacement
        MD_x_bootstrap = resample(MD_x, random_state=1234)
        kmeans_bootstrap = KMeans(n_clusters=k, n_init=10, random_state=1234)
        kmeans_bootstrap.fit(MD_x_bootstrap)
        boot_results[k].append(adjusted_rand_score(best_kmeans_models[k].labels_, kmeans_bootstrap.labels_))

# Plot adjusted Rand index for the bootstrapped clusters
mean_rand_indices = [np.mean(boot_results[k]) for k in clusters]
plt.plot(clusters, mean_rand_indices, marker='o')
plt.xlabel('Number of segments')
plt.ylabel('Adjusted Rand index')
plt.title('Bootstrapped Adjusted Rand Index')
plt.grid(True)
plt.show()

# Histogram for k=4
k4 = best_kmeans_models[4]
plt.hist(k4.labels_, bins=np.arange(5) - 0.5, density=True, edgecolor='black')
plt.xticks(range(4))
plt.xlabel('Cluster')
plt.ylabel('Frequency')
plt.title('Histogram for k=4')
plt.grid(True)
plt.show()

# Stability plot for k=4 (using silhouette scores as a proxy for stability)
from sklearn.metrics import silhouette_samples

silhouette_vals = silhouette_samples(MD_x, k4.labels_)
silhouette_avg = np.mean(silhouette_vals)

plt.bar(range(len(silhouette_vals)), silhouette_vals)
plt.axhline(y=silhouette_avg, color='r', linestyle='--')
plt.xlabel('Segment number')
plt.ylabel('Silhouette score')
plt.title('Segment Stability for k=4')
plt.ylim([0, 1])
plt.grid(True)
plt.show()

import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Define the path to your CSV file using a raw string
csv_file_path ='/content/drive/MyDrive/mcdonalds.csv'

# Load data from the CSV file using pandas
df = pd.read_csv(csv_file_path)

# Assuming the relevant columns are the first 11 columns
MD_x = np.array(df.iloc[:, 0:11] == "Yes", dtype=int)

# Function to calculate AIC, BIC, ICL for a range of k values
def fit_mixture_models(X, k_range):
    results = []
    for k in k_range:
        gmm = GaussianMixture(n_components=k, n_init=10, random_state=1234)
        gmm.fit(X)
        log_likelihood = gmm.score(X) * len(X)
        aic = gmm.aic(X)
        bic = gmm.bic(X)
        icl = bic + silhouette_score(X, gmm.predict(X)) * -2  # Proxy for ICL
        results.append((k, log_likelihood, aic, bic, icl, gmm))
    return results

# Fit models for k=2 to k=8
k_range = range(2, 9)
results = fit_mixture_models(MD_x, k_range)

# Extract the metrics
ks, log_likelihoods, aics, bics, icls, models = zip(*results)

# Plot the metrics
plt.figure(figsize=(10, 6))
plt.plot(ks, aics, label='AIC', marker='o')
plt.plot(ks, bics, label='BIC', marker='o')
plt.plot(ks, icls, label='ICL', marker='o')
plt.xlabel('Number of components')
plt.ylabel('Value of information criteria (AIC, BIC, ICL)')
plt.legend()
plt.grid(True)
plt.show()

# Select the model with 4 components
MD_m4 = models[ks.index(4)]

# Assuming you have the k-means model MD.k4 from the previous steps
k4 = best_kmeans_models[4]

# Create a contingency table
kmeans_clusters = k4.labels_
mixture_clusters = MD_m4.predict(MD_x)
contingency_table = pd.crosstab(kmeans_clusters, mixture_clusters)
print(contingency_table)


import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Define the path to your CSV file using a raw string
csv_file_path ='/content/drive/MyDrive/mcdonalds.csv'

# Load data from the CSV file using pandas
df = pd.read_csv(csv_file_path)

# Assuming the relevant columns are the first 11 columns
MD_x = np.array(df.iloc[:, 0:11] == "Yes", dtype=int)

# Perform K-Means Clustering
def kmeans_clustering(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=1234)
    kmeans.fit(X)
    return kmeans

# Assuming you have the best KMeans model from previous steps
kmeans_model = kmeans_clustering(MD_x, 4)
kmeans_labels = kmeans_model.labels_

# Perform Gaussian Mixture Model with clusters from K-Means
gmm = GaussianMixture(n_components=4, n_init=10, random_state=1234)
gmm.fit(MD_x)

# Create contingency table
gmm_labels = gmm.predict(MD_x)
contingency_table = pd.crosstab(kmeans_labels, gmm_labels)
print("Contingency Table:")
print(contingency_table)

# Calculate log-likelihoods
log_likelihood_m4a = gmm.score(MD_x) * len(MD_x)
print(f"Log Likelihood (MD.m4a): {log_likelihood_m4a}")

# Assuming we have the log-likelihood for the original model (MD.m4)
# For demonstration purposes, we'll use the same model as MD.m4
log_likelihood_m4 = log_likelihood_m4a
print(f"Log Likelihood (MD.m4): {log_likelihood_m4}")

# Transform and analyze 'Like' column
like_mapping = {
    "I hate it!": -5, "-5": -5, "-4": -4, "-3": -3, "-2": -2,
    "-1": -1, "0": 0, "+1": 1, "+2": 2, "+3": 3, "+4": 4, "I love it!": 5
}
df['Like.n'] = 6 - df['Like'].map(like_mapping)

# Print the transformed 'Like.n' table
like_table = df['Like.n'].value_counts().sort_index()
print("Transformed 'Like.n' Table:")
print(like_table)

# Create formula for regression
features = df.columns[:11]
formula = 'Like.n ~ ' + ' + '.join(features)
print("Regression Formula:")
print(formula)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Example data creation (replace with actual data)
np.random.seed(0)
# Assume MD_k4 contains cluster assignments for each observation
MD_k4 = pd.DataFrame(np.random.randint(1, 5, size=(100, 1)), columns=['Cluster'])
# Assume mcdonalds contains the necessary columns
mcdonalds = pd.DataFrame({
    'Like.n': np.random.rand(100),
    'Age': np.random.randint(18, 70, size=100),
    'VisitFrequency': np.random.choice(['Rarely', 'Often', 'Never'], size=100),
    'Gender': np.random.choice(['Male', 'Female'], size=100)
})

# Create a target variable where k4 == 3
target = (MD_k4['Cluster'] == 3).astype(int)

# Encode categorical variables
mcdonalds_encoded = pd.get_dummies(mcdonalds, drop_first=True)

# Train a decision tree classifier
features = mcdonalds_encoded[['Like.n', 'Age', 'VisitFrequency_Often', 'VisitFrequency_Rarely', 'Gender_Male']]
tree = DecisionTreeClassifier(random_state=0)
tree.fit(features, target)

# Plot the decision tree
plt.figure(figsize=(20, 10))
plot_tree(tree, feature_names=features.columns, class_names=['Not Cluster 3', 'Cluster 3'], filled=True)
plt.title('Decision Tree Classifier')
plt.show()
```

```python
import numpy as np
import pandas as pd

# Example data creation (replace with actual data)
np.random.seed(0)
# Assume MD_k4 contains cluster assignments for each observation
MD_k4 = pd.DataFrame(np.random.randint(1, 5, size=(100, 1)), columns=['Cluster'])
# Assume mcdonalds contains the necessary columns
mcdonalds = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], size=100)
})

# Concatenate cluster information with mcdonalds data
mcdonalds['Cluster'] = MD_k4['Cluster']

# Calculate the proportion of females for each cluster
female_proportion = mcdonalds.groupby('Cluster')['Gender'].apply(lambda x: (x == 'Female').mean())

# Print the results
print(female_proportion)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Example data creation (replace with actual data)
np.random.seed(0)
# Assume visit, like, and female_proportion are calculated as in the previous examples
visit = pd.Series([3.040426, 2.482490, 3.891975, 3.950249])
like = pd.Series([-0.1319149, -2.4902724, 2.2870370, 2.7114428])
female_proportion = pd.Series([0.5851064, 0.4319066, 0.4783951, 0.6144279])
cluster_labels = ['1', '2', '3', '4']

# Plot the scatter plot
plt.figure(figsize=(10, 7))
plt.scatter(visit, like, s=10 * female_proportion, alpha=0.7)

# Add text labels for each cluster
for i, label in enumerate(cluster_labels):
    plt.text(visit[i], like[i], label)

# Set axis limits and labels
plt.xlim(2, 4.5)
plt.ylim(-3, 3)
plt.xlabel('Visit Frequency')
plt.ylabel('Liking')

# Show the plot
plt.title('Visit Frequency vs Liking')
plt.grid(True)
plt.show()
```
