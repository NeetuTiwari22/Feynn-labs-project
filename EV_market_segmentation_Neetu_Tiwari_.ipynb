```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/content/cars_ds.csv')
df.head()

# Strip any leading/trailing spaces in column names
df.columns = df.columns.str.strip()

# Convert 'Ex-Showroom_Price' to numeric by removing 'Rs.' and commas, then to float
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)

# Convert 'Displacement' to numeric by removing 'cc' and commas, then to float
df['Displacement'] = df['Displacement'].str.replace(' cc', '').str.replace(',', '').astype(float)

# Distribution of Car Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Ex-Showroom_Price'].dropna(), bins=30, kde=True)
plt.title('Distribution of Ex-Showroom Prices')
plt.xlabel('Ex-Showroom Price (in INR)')
plt.ylabel('Frequency')
plt.show()

# Distribution of Car Displacement
plt.figure(figsize=(10, 6))
sns.histplot(df['Displacement'].dropna(), bins=30, kde=True)
plt.title('Distribution of Car Displacement')
plt.xlabel('Displacement (in cc)')
plt.ylabel('Frequency')
plt.show()

# Count of Cars by Make
plt.figure(figsize=(12, 8))
df['Make'].value_counts().plot(kind='bar')
plt.title('Count of Cars by Make')
plt.xlabel('Make')
plt.ylabel('Number of Cars')
plt.xticks(rotation=90)
plt.show()

# Count of Cars by Number of Cylinders
plt.figure(figsize=(10, 6))
sns.countplot(x='Cylinders', data=df)
plt.title('Count of Cars by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Number of Cars')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/content/Road-Accidents-2018-Table-2.4.csv'
data = pd.read_csv('/content/Road-Accidents-2018-Table-2.4.csv')

# Clean the data by removing rows with '% Share in Total'
clean_data = data[data['Vehicle Type'].notna() & ~data['Vehicle Type'].str.contains('% Share in Total')]

# Convert columns to appropriate data types using .loc
columns_to_convert = [
    '2017 - No. of Accidents', '2017 - Persons Killed', '2017 - Persons Injured',
    '2018 - No. of Accidents', '2018 - Persons Killed', '2018 - Persons Injured',
    '% Increase - No. of Accidents', '% Increase - Persons Killed', '% Increase - Persons Injured'
]
clean_data.loc[:, columns_to_convert] = clean_data.loc[:, columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Bar chart for number of accidents by vehicle type
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

clean_data.plot(kind='bar', x='Vehicle Type', y=['2017 - No. of Accidents', '2018 - No. of Accidents'], ax=ax[0])
ax[0].set_title('Number of Accidents by Vehicle Type (2017 vs 2018)')
ax[0].set_ylabel('Number of Accidents')

# Bar chart for number of persons killed by vehicle type
clean_data.plot(kind='bar', x='Vehicle Type', y=['2017 - Persons Killed', '2018 - Persons Killed'], ax=ax[1])
ax[1].set_title('Number of Persons Killed by Vehicle Type (2017 vs 2018)')
ax[1].set_ylabel('Number of Persons Killed')

# Bar chart for number of persons injured by vehicle type
clean_data.plot(kind='bar', x='Vehicle Type', y=['2017 - Persons Injured', '2018 - Persons Injured'], ax=ax[2])
ax[2].set_title('Number of Persons Injured by Vehicle Type (2017 vs 2018)')
ax[2].set_ylabel('Number of Persons Injured')

plt.tight_layout()
plt.show()

# Line chart for percentage increase
fig, ax = plt.subplots(figsize=(10, 6))

clean_data.plot(kind='line', x='Vehicle Type', y=['% Increase - No. of Accidents', '% Increase - Persons Killed', '% Increase - Persons Injured'], ax=ax)
ax.set_title('Percentage Increase in Accidents, Persons Killed, and Persons Injured (2017 to 2018)')
ax.set_ylabel('Percentage Increase')

plt.show()
```
